DataFrame –º–æ–∂–Ω–æ –∏–Ω–¥–µ–∫—Å–∏—Ä–æ–≤–∞—Ç—å –ø–æ-—Ä–∞–∑–Ω–æ–º—É. –î–ª—è –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –æ—Ç–¥–µ–ª—å–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏—é –≤–∏–¥–∞ `DataFrame["Name"]`. –î–ª—è –ª–æ–≥–∏—á–µ—Å–∫–æ–π –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–∏ ‚Äî `df[P(df["Name"])]`, –≥–¥–µ `P` ‚Äî —ç—Ç–æ –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ –ª–æ–≥–∏—á–µ—Å–∫–æ–µ —É—Å–ª–æ–≤–∏–µ, –ø—Ä–æ–≤–µ—Ä—è–µ–º–æ–µ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ —Å—Ç–æ–ª–±—Ü–∞ `"Name"`.

–í–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —ç—Ç–∏–º –¥–ª—è –æ—Ç–≤–µ—Ç–∞ –Ω–∞ –≤–æ–ø—Ä–æ—Å: –∫–∞–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –º–∞–≥–Ω–∏—è –≤ –∞–ª–∫–æ–≥–æ–ª–µ —Å –∫—Ä–µ–ø–æ—Å—Ç—å—é –Ω–∏–∂–µ $12\%$?

–ù–∞–ø—Ä–∏–º–µ—Ä, –º–µ—Ç–æ–¥ `map` –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –∑–∞–º–µ–Ω—ã –∑–Ω–∞—á–µ–Ω–∏–π –≤ –∫–æ–ª–æ–Ω–∫–µ, –ø–µ—Ä–µ–¥–∞–≤ –µ–º—É –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ —Å–ª–æ–≤–∞—Ä—å –≤–∏–¥–∞ `{old_value: new_value}`.

[[blog] ‚úèÔ∏è –õ—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü–∏–∏ –≤ Python](https://habr.com/ru/companies/piter/articles/674234/)

**Seaborn**

–¢–µ–ø–µ—Ä—å –¥–∞–≤–∞–π—Ç–µ –ø–µ—Ä–µ–π–¥–µ–º –∫ –±–∏–±–ª–∏–æ—Ç–µ–∫–µ `seaborn`. `Seaborn` ‚Äî –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–æ—É—Ä–æ–≤–Ω–µ–≤–æ–µ API –Ω–∞ –±–∞–∑–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ `matplotlib`. `Seaborn` —Å–æ–¥–µ—Ä–∂–∏—Ç –±–æ–ª–µ–µ –∞–¥–µ–∫–≤–∞—Ç–Ω—ã–µ –¥–µ—Ñ–æ–ª—Ç–Ω—ã–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤. –¢–∞–∫–∂–µ –≤ –±–∏–±–ª–∏–æ—Ç–µ–∫–µ –µ—Å—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Å–ª–æ–∂–Ω—ã–µ —Ç–∏–ø—ã –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –≤ `matplotlib` –ø–æ—Ç—Ä–µ–±–æ–≤–∞–ª–∏ –±—ã –±–æ–ª—å—à–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–¥–∞.

–ü–æ–∑–Ω–∞–∫–æ–º–∏–º—Å—è —Å –ø–µ—Ä–≤—ã–º —Ç–∞–∫–∏–º "—Å–ª–æ–∂–Ω—ã–º" —Ç–∏–ø–æ–º –≥—Ä–∞—Ñ–∏–∫–æ–≤ ‚Äî `pair plot` (`scatter plot matrix`). –≠—Ç–∞ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ–º–æ–∂–µ—Ç –Ω–∞–º –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –Ω–∞ –æ–¥–Ω–æ–π –∫–∞—Ä—Ç–∏–Ω–∫–µ, –∫–∞–∫ —Å–≤—è–∑–∞–Ω—ã –º–µ–∂–¥—É —Å–æ–±–æ–π —Ä–∞–∑–ª–∏—á–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏.

````python
import seaborn as sns

data, labels = load_wine(return_X_y=True, as_frame=True)
df = pd.concat([data, labels], axis=1)
cols = ["alcohol", "malic_acid", "ash", "target"]
sns_plot = sns.pairplot(df[cols])
sns_plot.savefig("pairplot.png")
````

–ö–∞–∫ –º–æ–∂–Ω–æ –≤–∏–¥–µ—Ç—å, –Ω–∞ –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –º–∞—Ç—Ä–∏—Ü—ã –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω—ã –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–∞. –û—Å—Ç–∞–ª—å–Ω—ã–µ –∂–µ –≥—Ä–∞—Ñ–∏–∫–∏ ‚Äî —ç—Ç–æ –æ–±—ã—á–Ω—ã–µ `scatter plot` –¥–ª—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –ø–∞—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

–î–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ —Ñ–∞–π–ª—ã —Å—Ç–æ–∏—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ `savefig`.

–í—ã–≤–µ–¥–∏—Ç–µ –∞–Ω–∞–ª–æ–≥–∏—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –ø–æ –∏–Ω—ã–º 5 –∫–æ–ª–æ–Ω–∫–∞–º.

–° –ø–æ–º–æ—â—å—é `seaborn` –º–æ–∂–Ω–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ `hist plot`. –î–ª—è –ø—Ä–∏–º–µ—Ä–∞ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ `color_intensity`. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ —Ç–∞–∫ —Ç–æ–∂–µ –º–æ–∂–Ω–æ –æ–±—Ä–∞—â–∞—Ç—å—Å—è –∫ –∫–æ–ª–æ–Ω–∫–∞–º.

````python
sns.histplot(df.color_intensity, kde=True)
plt.show()
````

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –Ω–∞ –¥–∏–∞–ø–∞–∑–æ–Ω –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö, –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è `box plot`. –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ç–æ, –∫–∞–∫ —Å–≤—è–∑–∞–Ω—ã –º–µ–∂–¥—É —Å–æ–±–æ–π 5 –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞–µ–º—ã—Ö –∫—Ä–µ–ø–æ—Å—Ç–µ–π –Ω–∞–ø–∏—Ç–∫–æ–≤ –∏ `flavanoids`.

````python
top_alcohol = (
    df.alcohol.value_counts().sort_values(ascending=False).head(5).index.values
)
sns.boxplot(
    y="alcohol", x="flavanoids", data=df[df.alcohol.isin(top_alcohol)], orient="h"
)
plt.show()
````

`Box plot` —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –∫–æ—Ä–æ–±–∫–∏ (–ø–æ—ç—Ç–æ–º—É –æ–Ω –∏ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è `box plot`), —É—Å–æ–≤ –∏ —Ç–æ—á–µ–∫ (–∏–Ω–∞—á–µ –µ–≥–æ –Ω–∞–∑—ã–≤–∞—é—Ç *—è—â–∏–∫ —Å —É—Å–∞–º–∏*). –ö–æ—Ä–æ–±–∫–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∏–Ω—Ç–µ—Ä–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π —Ä–∞–∑–º–∞—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, —Ç–æ –µ—Å—Ç—å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ 25% (Q1) –∏ 75% (Q3) –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–∏. –ß–µ—Ä—Ç–∞ –≤–Ω—É—Ç—Ä–∏ –∫–æ—Ä–æ–±–∫–∏ –æ–±–æ–∑–Ω–∞—á–∞–µ—Ç –º–µ–¥–∏–∞–Ω—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

–£—Å—ã –æ—Ç–æ–±—Ä–∞–∂–∞—é—Ç –≤–µ—Å—å —Ä–∞–∑–±—Ä–æ—Å —Ç–æ—á–µ–∫ –∫—Ä–æ–º–µ –≤—ã–±—Ä–æ—Å–æ–≤, —Ç–æ –µ—Å—Ç—å –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–µ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–ø–∞–¥–∞—é—Ç –≤ –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ ($Q1 - 1.5*IQR$, $Q3 + 1.5*IQR$), –≥–¥–µ $IQR = Q3 - Q1$ ‚Äî –∏–Ω—Ç–µ—Ä–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π —Ä–∞–∑–º–∞—Ö. –¢–æ—á–∫–∞–º–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –æ–±–æ–∑–Ω–∞—á–∞—é—Ç—Å—è –≤—ã–±—Ä–æ—Å—ã (outliers) ‚Äî —Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –≤–ø–∏—Å—ã–≤–∞—é—Ç—Å—è –≤ –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ –∑–Ω–∞—á–µ–Ω–∏–π, –∑–∞–¥–∞–Ω–Ω—ã–π —É—Å–∞–º–∏ –≥—Ä–∞—Ñ–∏–∫–∞.

[[wiki] üìö Box plot](https://en.wikipedia.org/wiki/Box_plot)

**–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Å–≤–æ–π —è—â–∏–∫ —Å —É—Å–∞–º–∏!** –î–ª—è —ç—Ç–æ–≥–æ –≤—ã–±–µ—Ä–∏—Ç–µ –∫–∞–∫—É—é-–Ω–∏–±—É–¥—å –ø–æ–¥–≤—ã–±–æ—Ä–∫—É –¥–∞–Ω–Ω—ã—Ö, –∫–æ—Ç–æ—Ä—É—é –º–æ–∂–Ω–æ –≤–∏–∑—É–∞–ª—å–Ω–æ –∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å.

````python
sns.boxplot(
    y="alcohol", x="color_intensity", data=df[df.alcohol.isin(top_alcohol)], orient="h"
)
plt.show()
````

–ü–æ—Å–ª–µ–¥–Ω–∏–π –≥—Ä–∞—Ñ–∏–∫, –∫–æ—Ç–æ—Ä—ã–π —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –≤ —ç—Ç–æ–º –∑–∞–¥–∞–Ω–∏–∏, ‚Äî —ç—Ç–æ `heat map`. –°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –∫—Ä–µ–ø–æ—Å—Ç–∏ –≤ 5 –±–∏–Ω–æ–≤ (–ø—Ä–∏–º–µ—Ä–Ω–æ —Ç–∞–∫–æ–π –∂–µ –ø–æ–¥—Ö–æ–¥ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –ø—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º) –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —á–∏—Å–ª–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ (`proanthocyanins`) –ø–æ –¥–≤—É–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º.

````python
df["alcoholGroup"] = pd.cut(df["alcohol"], bins=5)
````

````python
platform_genre_sales = (
    df.pivot_table(
        index="target",
        columns="alcoholGroup",
        values="proanthocyanins",
        aggfunc="sum",
        observed=False,
    )
    .fillna(0)
    .map(float)
)
sns.heatmap(platform_genre_sales, annot=True, fmt=".1f", linewidths=0.05)
plt.show()
````

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –∞–Ω–∞–ª–æ–≥–∏—á–Ω—É—é —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É. –ë—ã—Ç—å –º–æ–∂–µ—Ç, –≤—ã –æ–∂–∏–¥–∞–µ—Ç–µ –∫–∞–∫–∏–µ-—Ç–æ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏?

````python
df["color_intensity_group"] = pd.cut(df["color_intensity"], bins=4)
tmp = (
    df.pivot_table(
        index="target",
        columns="color_intensity_group",
        values="alcohol",
        aggfunc="median",
        observed=False,
    )
    .fillna(0)
    .map(float)
)
sns.heatmap(tmp, annot=True, fmt=".1f", linewidths=0.05)
plt.show()
````

–°–æ–∑–¥–∞–π—Ç–µ –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü `y_pred_05`, –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –∫–ª–∞—Å—Å–∞ —Å –ø–æ—Ä–æ–≥–æ–º 0.5. –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –≤ `y_score=0.6`, —Ç–æ —ç—Ç–æ –∫–ª–∞—Å—Å 1, –µ—Å–ª–∏ –≤ `y_score=0.3`, —Ç–æ —ç—Ç–æ –∫–ª–∞—Å—Å 0.

–¢–µ–ø–µ—Ä—å —É –Ω–∞—Å –µ—Å—Ç—å –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è `precision` –∏ `recall` –ø—Ä–∏ –≤—Å–µ—Ö –ø–æ—Ä–æ–≥–∞—Ö. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ `f1-score` –∏ –≤—ã–±–µ—Ä–∏—Ç–µ –ø–æ—Ä–æ–≥, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º `f1-score` –º–∞–∫—Å–∏–º–∞–ª–µ–Ω:

````python
import numpy as np

f1 = 2*precision*recall/(precision+recall)

best_f1_index = np.argmax(f1)
best_threshold = thresholds[best_f1_index]

print("Best threshold", thresholds[best_f1_index])
````

````python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
````

–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ `cross_val_score` –∏–ª–∏ `cross_validate`:

````python
iris = load_iris()
x, y = iris.data, iris.target

model = KNeighborsClassifier(n_neighbors=3, metric="euclidean")

cv = KFold(n_splits=5, shuffle=True, random_state=42)
results = cross_val_score(model, x, y, cv=cv, scoring='accuracy')

print(results.mean())
````

–¢–µ–ø–µ—Ä—å —Å–¥–µ–ª–∞–π—Ç–µ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é, –∏—Å–ø–æ–ª—å–∑—É—è `KFold` –∏–∑ `sklearn.model_selection`:

````python
np.random.seed(42)

def kfold_cv(num_folds, x, y, model, metric=accuracy_score,random_state=42,shuffle = True):

    results = []
    kf = KFold(n_splits = num_folds,shuffle = shuffle,random_state=random_state)
    for i, (train_indices, val_indices) in enumerate(kf.split(x)):

        x_train = x[train_indices]
        y_train = y[train_indices]
        x_val = x[val_indices]
        y_val = y[val_indices]
        model.fit(x_train, y_train)
        predictions = model.predict(x_val)
        score = metric(y_val, predictions)
        results.append(score)
        print(f"Accuracy for fold {i + 1}: {score}")

    return np.array(results)

results = kfold_cv(5, x, y, model)
print("Mean: ", results.mean())
````

–ù–∞–ø–∏—à–∏—Ç–µ —Å–≤–æ—é —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é:

````python
def custom_cv(num_folds, x, y, model, metric=accuracy_score,shuffle = True,random_state=42):

    np.random.seed(random_state)
    results = []
    indices = np.arange(len(x))
    if shuffle:
        np.random.shuffle(indices)
    delta_i = x.shape[0]//num_folds
    for i in range(num_folds):
        val_indices = indices[i*delta_i:(i+1)*delta_i]
        train_indices = np.concatenate([indices[:i*delta_i],indices[(i+1)*delta_i:]])
        x_val = x[val_indices]
        y_val = y[val_indices]

        x_train = x[train_indices]
        y_train = y[train_indices]

        model.fit(x_train, y_train)
        predictions = model.predict(x_val)
        score = metric(y_val, predictions)
        results.append(score)
        print(f"Accuracy for fold {i + 1}: {score}")

    return np.array(results)


results = custom_cv(5, x, y, model)
print("Mean: ", results.mean())
````

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –º—ã —É–º—ã—à–ª–µ–Ω–Ω–æ –∑–∞–≥—Ä—É–∂–∞–µ–º –º–∞–ª—É—é —á–∞—Å—Ç—å –¥–∞—Ç–∞—Å–µ—Ç–∞ (10%) –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –±—ã—Å—Ç—Ä–µ–µ –¥–µ–ª–∞—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏—è. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ `data` –∏ `labels` –∫–∞–∫ –∏—Å—Ö–æ–¥–Ω—ã–µ `x` –∏ `y`.*

–§–æ—Ä–º—É–ª–∞ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ L2-—Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è (–∏–ª–∏ –µ–≤–∫–ª–∏–¥–æ–≤–æ–π –¥–∏—Å—Ç–∞–Ω—Ü–∏–∏) –º–µ–∂–¥—É –¥–≤—É–º—è –≤–µ–∫—Ç–æ—Ä–∞–º–∏  $a$ –∏ $b$ –≤—ã–≥–ª—è–¥–∏—Ç —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$$ d(a, b) = \sqrt{\sum_{i=1}^{D} (a_i - b_i)^2},$$

–≥–¥–µ:
- $ a $ –∏ $ b $ ‚Äî —ç—Ç–æ –≤–µ–∫—Ç–æ—Ä—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–≤—É—Ö —Ç–æ—á–µ–∫ (–∏–ª–∏ –æ–±—ä–µ–∫—Ç–æ–≤),
- $ D $ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–∏–ª–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞),
- $ a_i $ –∏ $ b_i $ ‚Äî —ç—Ç–æ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤–µ–∫—Ç–æ—Ä–∞ $ a $ –∏ $ b $ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –Ω–∞ –ø–æ–∑–∏—Ü–∏–∏ $ i $.


–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å k-NN –∏ —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ –µ–≥–æ –º–µ—Ç–æ–¥—ã.

–û—Ç –≤–∞—Å —Ç—Ä–µ–±—É–µ—Ç—Å—è —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ `compute_distances`, –∫–æ—Ç–æ—Ä—ã–π –¥–æ–ª–∂–µ–Ω –≤—ã—á–∏—Å–ª—è—Ç—å –º–∞—Ç—Ä–∏—Ü—É –ø–æ–ø–∞—Ä–Ω—ã—Ö –µ–≤–∫–ª–∏–¥–æ–≤—ã—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É –º–∞—Å—Å–∏–≤–æ–º –ø–æ–¥–∞–≤–∞–µ–º—ã—Ö –Ω–∞ –≤—Ö–æ–¥ –æ–±—ä–µ–∫—Ç–æ–≤ `test` –∏ –º–∞—Å—Å–∏–≤–æ–º –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏ `self.train_data`. –ü–æ—Å—Ç–∞—Ä–∞–π—Ç–µ—Å—å —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ä–∞—Å—á–µ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ü–∏–∫–ª–æ–≤.

–¢–∞–∫–∂–µ –æ—Ç –≤–∞—Å —Ç—Ä–µ–±—É–µ—Ç—Å—è —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ `predict`, –∫–æ—Ç–æ—Ä—ã–π:
* –¥–ª—è –ø–æ–¥–∞–≤–∞–µ–º–æ–≥–æ –Ω–∞ –≤—Ö–æ–¥ –º–∞—Å—Å–∏–≤–∞ –æ–±—ä–µ–∫—Ç–æ–≤ `x` –ø—Ä–æ–≤–µ–¥–µ—Ç —Ä–∞—Å—á–µ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –¥–æ –æ–±—ä–µ–∫—Ç–æ–≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏,
* –æ–ø—Ä–µ–¥–µ–ª–∏—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –∏–∑ `x` —Ç–æ–ø-k –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π –∏–∑ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏ —Å–æ–≥–ª–∞—Å–Ω–æ —Ä–∞—Å—á–∏—Ç–∞–Ω–Ω—ã–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è–º (–ø—Ä–∏–≥–æ–¥–∏—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è `np.argsort` [üõ†Ô∏è[doc]](https://numpy.org/doc/stable/reference/generated/numpy.argsort.html)),
* —Å—Ñ–æ—Ä–º–∏—Ä—É–µ—Ç –∏—Ç–æ–≥–æ–≤—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –æ–±—ä–µ–∫—Ç–æ–≤ –∏–∑ `x`, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ –º–µ—Ç–∫–∞—Ö –∫–ª–∞—Å—Å–æ–≤ –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π (–ø—Ä–∏–≥–æ–¥–∏—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è `scipy.stats.mode` [üõ†Ô∏è[doc]](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mode.html)).

````python
from scipy.stats import mode


class kNN:
    def __init__(self, k):
        self.k = k

    def fit(self, x, y):
        self.train_data = x.copy()
        self.train_labels = y.copy()

    def distance_func(self,x,y):
      return np.sqrt(np.sum((x-y)**2))


    def compute_distances(self, test):

        distances = np.array([[self.distance_func(x,y) for x in self.train_data]for y in test])

        return distances

    def predict(self, x):
        distances = self.compute_distances(x)
        # top k idx
        indexes = np.argsort(distances, axis=1)[:, :self.k]
        labels_of_top_classes = self.train_labels[indexes]
        predicted_class, _ = mode(labels_of_top_classes, axis=1)
        return predicted_class
````

````python
from sklearn.model_selection import  GridSearchCV

knn = KNeighborsClassifier()

model =  GridSearchCV(
    estimator=knn,
    cv=KFold(3, shuffle=True, random_state=42),
    param_grid={
        "n_neighbors": np.arange(1, 20),
        "metric": ["euclidean", "manhattan"],
    },
    scoring="accuracy",
    n_jobs=-1,
)
model.fit(x_train, y_train)

print('–û–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Å–µ–¥–µ–π, –Ω–∞–π–¥–µ–Ω–Ω–æ–µ —Å –ø–æ–º–æ—â—å—é RandomizedSearchCV',
      model.best_params_["n_neighbors"])
````

–í —Å–ª—É—á–∞–µ multilabel:  –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –º–∞—Ç—Ä–∏—Ü –æ—à–∏–±–æ–∫ (confusion matrix) –∏ —Ä–∞—Å—Å—á—ë—Ç–æ–≤ –º–µ—Ç—Ä–∏–∫ –º–æ–∂–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Å–ª–µ–¥—É—é—â–∏–º –∫–æ–¥–æ–º.

````python
def calculate_metrics(confusion_matrix):
    # Select TP, FP, TN, FN
    TP = confusion_matrix[1, 1]  # True Positive
    FP = confusion_matrix[0, 1]  # False Positive
    TN = confusion_matrix[0, 0]  # True Negative
    FN = confusion_matrix[1, 0]  # False Negative

    # Calculate precision, recall, F1-score
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1_score = (
        (2 * precision * recall / (precision + recall))
        if (precision + recall) > 0
        else 0
    )

    return precision, recall, f1_score
````

````python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4), sharey="row")

true_values = [0, 1, 1, 0, 0, 0, 0, 1]
predicted_values = [1, 1, 9, 0, 0, 1, 0, 1]

confusion_matrices = multilabel_confusion_matrix(true_values, predicted_values)
precisions, recalls, f1_scores = [], [], []
for i, confusion_matrix in enumerate(confusion_matrices):
    disp = ConfusionMatrixDisplay(confusion_matrix)
    disp.plot(ax=axes[i])

    precision, recall, f1_score = calculate_metrics(confusion_matrix)
    f1_scores.append(f1_score)
    precisions.append(precision)
    recalls.append(recall)

plt.show()
````

````python
print(f"Mean Precision: {np.array(precisions).mean()}")
print(f"Mean Recall: {np.array(recalls).mean()}")
print(f"Mean F1-score: {np.array(f1_scores).mean()}")
````

### –ë–æ–ª–µ–Ω-–∑–¥–æ—Ä–æ–≤

–†–µ—à–∞–µ–º –∑–∞–¥–∞—á–∏ –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏. –î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–µ–º classification_report + –≥—Ä–∞—Ñ–∏–∫–∏ (PR-–∫—Ä–∏–≤–∞—è, ROC-–∫—Ä–∏–≤–∞—è) –∏ confusion_matrix. –í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –±—É–¥–µ–º —Ä–∞–±–æ—Ç–∞—Ç—å —Å–æ –≤—Å–µ–º–∏ –¥–∞–Ω–Ω—ã–º–∏ (—Ç.–µ —Å df, –∞ –Ω–µ —Å df_short). –ü–æ –¥–∞–Ω–Ω—ã–º –≤–∏–¥–Ω–æ, —á—Ç–æ —Å —ç—Ç–æ–π –∑–∞–¥–∞—á–µ–π —Ö—É–∂–µ –≤—Å–µ–≥–æ —Å–ø—Ä–∞–≤–ª—è–µ—Ç—Å—è –º–æ–¥–µ–ª—å 3. –ú–æ–¥–µ–ª—å 1 –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –ª—É—á—à–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –Ω–∞ PR-–∫—Ä–∏–≤–æ–π, –∞ –º–æ–¥–µ–ª—å 2 - –Ω–∞ ROC-–∫—Ä–∏–≤–æ–π. –ü—Ä–∏ —ç—Ç–æ–º –∏—Å—Ö–æ–¥—è –∏–∑ confusion_matrix –º–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ –º–æ–¥–µ–ª—å 2 —Ä–µ–∂–µ –æ—Ç–Ω–æ—Å–∏—Ç –±–æ–ª—å–Ω—ã—Ö –ª—é–¥–µ–π –∫ –∑–¥–æ—Ä–æ–≤—ã–º.

–í –∑–∞–¥–∞—á–∞—Ö –º–µ–¥–∏—Ü–∏–Ω—ã –Ω–∞–º –≤–∞–∂–Ω–µ–µ –Ω–µ –ø—Ä–æ–ø—É—Å—Ç–∏—Ç—å –±–æ–ª—å–Ω–æ–≥–æ —á–µ–ª–æ–≤–µ–∫–∞, –ø–æ—ç—Ç–æ–º—É –≤–∞–∂–Ω–µ–µ —É—á–∏—Ç—ã–≤–∞—Ç—å recall. –í –∫–∞—á–µ—Å—Ç–≤–µ –∏—Ç–æ–≥–æ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å $F_{\beta}-score$, –≥–¥–µ $\beta = 2$, —á—Ç–æ–±—ã –±–æ–ª—å—à–µ —É—á–∏—Ç—ã–≤–∞—Ç—å recall. –ü–æ —ç—Ç–æ–π –º–µ—Ç—Ä–∏–∫–µ –º–æ–¥–µ–ª—å 2 —Å—Ä–∞–±–æ—Ç–∞–ª–∞ –ª—É—á—à–µ.

````python
from sklearn.metrics import precision_recall_curve, auc, roc_curve, roc_auc_score,confusion_matrix,classification_report,fbeta_score

model_N = [1,2,3]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

for model in model_N:
  precision, recall, thresholds = precision_recall_curve(df["–ó–¥–æ—Ä–æ–≤"], df[f'–ó–¥–æ—Ä–æ–≤_{model}'])
  pr_auc = auc(recall, precision)
  axes[0].plot(recall, precision, label=f"PR curve (area = {pr_auc:.2f}), model #{model}")

axes[0].set_xlabel("Recall")
axes[0].set_ylabel("Precision")
axes[0].set_title("Precision-Recall Curve")
axes[0].legend(loc="lower left")

for model in model_N:
  fpr, tpr, thresholds = roc_curve(df["–ó–¥–æ—Ä–æ–≤"], df[f'–ó–¥–æ—Ä–æ–≤_{model}'])
  roc_auc = roc_auc_score(df["–ó–¥–æ—Ä–æ–≤"], df[f'–ó–¥–æ—Ä–æ–≤_{model}'])
  axes[1].plot(fpr, tpr, label=f"ROC curve (area = {roc_auc:.2f})), model #{model}")
  axes[1].plot([0, 1], [0, 1], 'k--')

axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].set_title("ROC Curve")
axes[1].legend(loc='lower right')
plt.show()

fig,axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4),sharey='row')
for model in model_N:
  conf_mat = confusion_matrix(df["–ó–¥–æ—Ä–æ–≤"],df[f'–ó–¥–æ—Ä–æ–≤_{model}'])
  disp = ConfusionMatrixDisplay(conf_mat)
  disp.plot(ax=axes[model-1])
  axes[model-1].set_title(f'model #{model}')
plt.show()

target_names = ["–ë–æ–ª–µ–Ω", "–ó–¥–æ—Ä–æ–≤"]
for model in model_N:
  print(f'model #{model}')
  print(classification_report(df["–ó–¥–æ—Ä–æ–≤"], df[f"–ó–¥–æ—Ä–æ–≤_{model}"], target_names=target_names,digits=3))
  classification_report_dict = classification_report(df["–ó–¥–æ—Ä–æ–≤"], df[f"–ó–¥–æ—Ä–æ–≤_{model}"], target_names=target_names, output_dict=True,digits = 4)
  model_metrics_data[f'sick_or_healthy_model{model}'] = classification_report_dict['macro avg']['f1-score']
  print('f2_score',fbeta_score(df["–ó–¥–æ—Ä–æ–≤"], df[f"–ó–¥–æ—Ä–æ–≤_{model}"],beta=2))
  print('################')
````

### –î–æ–∑–∞ –ª–µ–∫–∞—Ä—Å—Ç–≤–∞

–†–µ—à–∞–µ–º –∑–∞–¥–∞—á—É —Ä–µ–≥—Ä–µ—Å—Å–∏. –û—Ü–µ–Ω–∏–≤–∞–µ–º mse –∏ mae. –ü–æ –≥—Ä–∞—Ñ–∏–∫–∞–º mse –≤–∏–¥–Ω–æ, —á—Ç–æ —É –º–æ–¥–µ–ª–∏ 2 mse –≤—ã—à–µ –¥–ª—è –≤—Å–µ—Ö "–∏–Ω–¥–µ–∫—Å–æ–≤" –¥–æ–∑, —á–µ–º —É –¥–≤—É—Ö –¥—Ä—É–≥–∏—Ö. –ü–æ –∑–Ω–∞—á–µ–Ω–∏—è–º mse —É –º–æ–¥–µ–ª–∏ 3 –≤ —Å—Ä–µ–¥–Ω–µ–º –Ω–∏–∂–µ, —á–µ–º –º–æ–¥–µ–ª–∏ 1. –í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –æ–Ω–∞ –ª—É—á—à–µ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ä–µ–≥—Ä–µ—Å—Å–∏–∏.

````python
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_true = np.vstack(np.array(df_short["–î–æ–∑–∏—Ä–æ–≤–∫–∞"],dtype=object))

fig,axes = plt.subplots(nrows=1, ncols=len(model_N), figsize=(12, 4),sharey='row')
target_names = [str(name) for name in list(range(1,4))]
for model in model_N:
  y_pred = np.vstack(np.array(df_short[f"–î–æ–∑–∏—Ä–æ–≤–∫–∞_{model}"],dtype=object))
  mse = mean_squared_error(y_true,y_pred,multioutput='raw_values')
  mae = mean_absolute_error(y_true, y_pred,multioutput='raw_values')
  print(f'model #{model}: average_mse = {mse.mean():.3f}, average mae = {mae.mean():.3f}')
  model_metrics_data[f'dosage_model{model}_mse'] = 1-((mse - mse.min()) / (mse.max() - mse.min())).mean()
  print(f'(1-mse_norm) -',model_metrics_data[f'dosage_model{model}_mse'])
  sns.barplot(x=target_names, y=mse,ax = axes[model-1],color='hotpink')
  axes[model-1].set_title(f'model #{model}')
  axes[model-1].set_xlabel("dosage index")
  axes[model-1].set_ylabel("mse")

````

### –ò—Ç–æ–≥


–ß—Ç–æ–±—ã –æ—Ü–µ–Ω–∏—Ç—å, –∫–∞–∫–∞—è –º–æ–¥–µ–ª—å –≤ –∏—Ç–æ–≥–µ —Å—Ä–∞–±–æ—Ç–∞–ª–∞ –ª—É—á—à–µ –¥—Ä—É–≥–∏—Ö, –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤–∑–≤–µ—à–µ–Ω–Ω—É—é —Å—É–º–º—É —á–µ—Ç—ã—Ä–µ—Ö –º–µ—Ç—Ä–∏–∫. –°–∞–º—ã–π –±–æ–ª—å—à–æ–π –≤–µ—Å –∏–º–µ–µ—Ç –ø–µ—Ä–≤–∞—è –ø–æ–¥–∑–∞–¥–∞—á–∞, –Ω–∞–¥–æ —É–º–µ—Ç—å –Ω–∞—Ö–æ–¥–∏—Ç –±–æ–ª—å–Ω—ã—Ö –ª—é–¥–µ–π. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ç–∏–ø–∞ –±–æ–ª–µ–∑–Ω–∏ –∏ —Ç–∏–ø–∞ –ª–µ–∫–∞—Ä—Å—Ç–≤–∞ –∏–º–µ—é—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—ã–π –≤–µ—Å. –Ø –¥—É–º–∞—é, —á—Ç–æ –æ–¥–∏–Ω–∞–∫–æ–≤–æ –≤–∞–∂–Ω–æ –≤–µ—Ä–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Ç–∏–ø –±–æ–ª–µ–∑–Ω–∏ –∏ —Ç–∏–ø –ª–µ–∫–∞—Ä—Å—Ç–≤–∞, –ø–æ—Ç–æ–º—É —á—Ç–æ –ª–µ—á–∏—Ç—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é –±–æ–ª–µ–∑–Ω—å –Ω–µ —Ç–µ–º –ª–µ–∫–∞—Ä—Å—Ç–≤–æ–º –º–æ–∂–µ—Ç –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –ø–ª–æ—Ö–æ–º—É –∏—Å—Ö–æ–¥—É. –ú–µ–Ω–µ–µ –≤–∞–∂–Ω–∞ –≤ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞—Ö –¥–æ–∑–∏—Ä–æ–≤–∫–∞. –ï—Å–ª–∏ —á—É—Ç—å-—á—É—Ç—å –ø–µ—Ä–µ–±–æ—Ä—â–∏–ª–∏ –∏–ª–∏ –Ω–µ –¥–æ–ª–æ–∂–∏–ª–∏, –æ–±—â–µ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –Ω–µ –≤—Å–µ–≥–¥–∞ –º–æ–∂–µ—Ç —É—Ö—É–¥—à–∏—Ç—å—Å—è.

````python
def composite_metric(model_numb,weights = [0.4,0.25,0.25,0.1]):
  return (weights[0]*model_metrics_data[f'sick_or_healthy_model{model_numb}']+
          weights[1]*model_metrics_data[f'sickness_type_model{model_numb}']+
          weights[2]*model_metrics_data[f'medicine_model{model_numb}']+
          weights[3]*model_metrics_data[f'dosage_model{model_numb}_mse'])

for model in model_N:
  final_metric = composite_metric(model)
  print(f'model #{model} composite metric: {final_metric:.3f}')
````
